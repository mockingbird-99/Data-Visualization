## Dataset Overview

> The data set explored in this project is a 2007 flight data. 
The data set is large with 7.5 million rows and 29 columns. 
It comtains 7 million instances of data on 2007 flights for 
features such as the month, day of the month, day of the week, 
flight carrirer (airline), destination, and origin. 
The data set can be obtained from 
[Havard Dataverse](https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/HG7NV7#)

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
dff = pd.read_csv('dff.csv')
```

> The data was wrangled in 2 steps to clean the data
1. Duplicates were dropped
2. Month, DayofMonth, DayOfWeek and Diverted variables were converted from int data types to objects

## Distribution of Flights amongst Airlines

> 20 different airlines are captured in this data, and the 7.5 Million flights are shared amongst them. The distribution is studied in the following visualization. The barchart of relative frequencies is depicted to show the distribution. Almost 16% of all flights are carried by WN airlines, followed by AA and OO at 9% and 8.5% respectively HA and AQ each have less than 1%

```python
#select first color from seaborn color palette
base_color = sb.color_palette()[0]
order_u = dff.UniqueCarrier.value_counts().index

#function for barchart plotting
def sbplot(feature,order):
    sb.countplot(data=dff, y=feature, color=base_color, order=order);


#Plot relative frequencies of Unique Carriers
order_un = dff.UniqueCarrier.value_counts()
sum_unique = dff.UniqueCarrier.value_counts().sum()
max_unique = order_un[0]
max_prop = max_unique/sum_unique

#create list of string values for for tick labels 
tick_props = np.arange(0,max_prop+0.02, 0.01)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]

plt.figure(figsize=[14.70, 8.27]) #Adjust figure size
sbplot(dff.UniqueCarrier,order_u)
plt.title('Count of flights for each Carrier')
plt.xticks(tick_props*sum_unique, tick_names)
plt.xlabel('Proportions for airlines');
```

## Airlines with the most cancelled flights

> Analysing the cancelled flights, we can study which airlines have more cancelled flights. The visual depicts that even though WN airlines have the highest customers, they have one of the least number of cancelled flights. MQ, YV and OH airlines have many cancelled flights.

```python
#barplot
plt.figure(figsize=[14.70, 8.27])
sb.barplot(data=dff, x='Cancelled', y='UniqueCarrier', color=base_color)
plt.title('Bar plot of Cancelled flights and Flight Carriers');
```

## (Visualization 3)

>There were 4 different causes of delay captured in the data set, namely Late Aircraft Delay, NAS Delay, Weather Delay, Security Delay, and Carrier Delay. Since the flight carrier are the key variables of this study, Carrier Delay is visualized with the flight carriers. It is observed that EV airline has the highest number of carrier delays. It is worthy of note that WN which has the highest number of flights has one of the lowest number of carrier delays.


```python
#barplot
plt.figure(figsize=[14.70, 8.27])
sb.barplot(data=dff, x='CarrierDelay', y='UniqueCarrier', color=base_color)
plt.title('Bar plot of Carrier Delay and Flight Carriers');
```
